# Descriptive Statistics

1. A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?
-  The most representative value is around 12 dozens because the average (mean) of the data is close to 12, which best represents weekly sales.

2. A teacher records the marks of her students in a short quiz: [12, 15, 14, 16, 18, 20, 19].
What is the mean score, and what does it tell us about the class’s performance?
- The mean score is about 16.3. This shows that the class performed well on average.

3. A store records the shoe sizes sold in one day: [7, 8, 9, 8, 8, 10, 7, 9].
What is the mode, and why is this information useful for the store manager?
- The mode is 8 because it appears most frequently. This helps the store manager stock more of size 8 shoes

4. A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].
Why is the median a better measure than the mean in this case? Calculate the median.
- The median is 10,200. It is better than the mean because the value 50,000 is an outlier and can distort the average

5. A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40].
What does the range tell us about the variation in the student’s puzzle-solving time?
- The range is 15 minutes. This shows that the student’s puzzle-solving time varies by 15 minutes.

6. A farmer records the weekly weight of harvested apples (kg): [100, 105, 98, 110, 120].
Find the range. How can this help the farmer in planning his packaging?
- The range is 22 kg. This helps the farmer understand variation in weights for better packaging

7. Two delivery companies track delivery delays (in minutes).
Company A: variance = 6
Company B: variance = 15
 Which company is more consistent, and why?
- Company A is more consistent because it has lower variance

8. A finance student compares the daily price fluctuations of two cryptocurrencies.
Coin A: standard deviation = $30
Coin B: standard deviation = $120
 Which coin is riskier to invest in, and why
- Coin B is riskier because it has a higher standard deviation, meaning more fluctuation.

9. A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].
- The mean electricity usage is 414 kWh and standard deviation is about 20.7. This shows usage is fairly consistent.

10. A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21].
Find the mean, median, mode, range, and standard deviation. What insights can these measures provide
about the player’s scoring performance?
- Mean = 19.6, Median = 19.5, Mode = none, Range = 10, Standard deviation ≈ 3.16. This shows the player is consistent with moderate variation.
